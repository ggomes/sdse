```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data = datasets.load_iris(as_frame=True)
df = data['frame']
target_names = data['target_names'] # ['setosa', 'versicolor', 'virginica']
feature_names = data['feature_names'] # ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
```

```python
def make_plot(x1,x2):

    markers = ['^','.','+']
    markercolors = ['r','magenta','g']

    _, ax = plt.subplots(figsize=(4,4))
    ax.set_xlabel(feature_names[x1])
    ax.set_ylabel(feature_names[x2])

    for k, name in enumerate(target_names):
        ind = df['target']==k
        ax.scatter(df.loc[ind,feature_names[x1]],df.loc[ind,feature_names[x2]],
                s=20,
                color=markercolors[k],
                marker=markers[k],
                label=name)
    ax.legend()

    return ax
```

```python
ax = make_plot(2,3)
```
